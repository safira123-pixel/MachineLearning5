```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ms = pd.read_csv('mushrooms.csv')
ms.head()
```

```python
ms.info()
```

```python
from sklearn.preprocessing import LabelEncoder

# Inisiasi label encoder
encode = LabelEncoder()

# Terpakan label encoder
ms['class'] = encode.fit_transform(ms['class'])
ms['cap-shape'] = encode.fit_transform(ms['cap-shape'])
ms['cap-surface'] = encode.fit_transform(ms['cap-surface'])
ms['cap-color'] = encode.fit_transform(ms['cap-color'])
ms['bruises'] = encode.fit_transform(ms['bruises'])
ms['odor'] = encode.fit_transform(ms['odor'])
ms['gill-attachment'] = encode.fit_transform(ms['gill-attachment'])
ms['gill-spacing'] = encode.fit_transform(ms['gill-spacing'])
ms['gill-size'] = encode.fit_transform(ms['gill-size'])
ms['gill-color'] = encode.fit_transform(ms['gill-color'])
ms['stalk-shape'] = encode.fit_transform(ms['stalk-shape'])
ms['stalk-root'] = encode.fit_transform(ms['stalk-root'])
ms['stalk-surface-above-ring'] = encode.fit_transform(ms['stalk-surface-above-ring'])
ms['stalk-surface-below-ring'] = encode.fit_transform(ms['stalk-surface-below-ring'])
ms['stalk-color-above-ring'] = encode.fit_transform(ms['stalk-color-above-ring'])
ms['stalk-color-below-ring'] = encode.fit_transform(ms['stalk-color-below-ring'])
ms['veil-type'] = encode.fit_transform(ms['veil-type'])
ms['veil-color'] = encode.fit_transform(ms['veil-color'])
ms['ring-number'] = encode.fit_transform(ms['ring-number'])
ms['ring-type'] = encode.fit_transform(ms['ring-type'])
ms['spore-print-color'] = encode.fit_transform(ms['spore-print-color'])
ms['population'] = encode.fit_transform(ms['population'])
ms['habitat'] = encode.fit_transform(ms['habitat'])

ms.head()
```

```python
from sklearn.model_selection import train_test_split
y = ms["class"].values
x = ms.drop(["class"], axis=1).values
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=6, random_state=1)

dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)
print(y_pred[0:5])
```

```python
from sklearn.metrics import accuracy_score

# Memprediksi test set labels
y_pred = dt.predict(X_test)

#  menghitung set accuracy
acc = accuracy_score(y_test, y_pred)
print("Test set accuracy: {:.2f}".format(acc))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Instantiate dt_entropy, set 'entropy' sebagai kriteria informasi
dt_entropy = DecisionTreeClassifier(max_depth=8, criterion='entropy', random_state=1)

# Fit dt_entropy kedalam training set
dt_entropy.fit(X_train, y_train)
```

```python
dt_gini = DecisionTreeClassifier(max_depth=8, criterion='gini', random_state=1)
dt_gini.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

# menggunakan dt_entropy untuk memprediksi test set labels
y_pred = dt_entropy.predict(X_test)
y_pred_gini = dt_gini.predict(X_test)

# mengevaluasi accuracy_entropy
accuracy_entropy = accuracy_score(y_test, y_pred)
accuracy_gini = accuracy_score(y_test, y_pred_gini)

# Print accuracy_entropy
print("Accuracy achieved by using entropy: ", accuracy_entropy)

# Print accuracy_gini
print("Accuracy achieved by using gini: ", accuracy_gini)
```
